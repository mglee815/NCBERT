```python
#데이터를 불러와서 사용하세요
#이 코드에서는 ISDS 학생들이 직접 레이블링한 카카오페이 관련 기사 댓글을 사용


#data = pd.read_excel("./감성분석을 원하는 데이터.csv") # load data

data = pd.read_csv("./comments_unseen.tsv", sep = '\t') # load data
data = data[['comments', 'sentiment']]
```

```python
# 실제 사용할때는 label된 sentiment columns가 없겠죠?
# 이번 예시에서는 정확도 체크를 위해 sentiment가 레이블 되어있는 데이터를 사용했습니다.
data 
```

```python
#미리 만들어둔 NCBERT 파이썬 모듈을 import해서 사용하면 됩니다.
#자세한 코드가 혹시 궁금하신분은 NCBERT_sentiment.py 파일을 살펴보시면 됩니다.

import NCBERT_sentiment as NCsenti
```

```python
NC_pred = NCsenti.main(data['comments']) # main함수에 list 형태로 텍스트를 넣으면 예측값을 return 해줍니다.
data['pred'] = NC_pred
```

```python
import numpy as np
np.mean(NC_pred == data['sentiment']) # 정확도는 0.7 정도 나오네요
```
